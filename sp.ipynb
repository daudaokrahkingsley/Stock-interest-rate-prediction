```python
## CREATING TESTING AND TRAINING DATASET/DATA CLEANING
```

```python
y =stock ['S&P 500 Price']
y
```

```python
x = stock[['Interest Rates', 'Employment']]
x
```

```python
#using 80% of the Data for Training and 20% for Testing. Using SK LEARN
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size =0.2)
```

```python
X_train
```

```python
from sklearn.model_selection import train_test_split
```

```python
y_test
```

```python
y_train
```

```python
# Model Testing and EValuation
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression(fit_intercept = True)
```

```python
regressor.fit(X_train, y_train)
```

```python
print('Linear Model Coefficies (m)', regressor.coef_)
```

```python
print('Linear Model Coefficies (m)', regressor.intercept_)
```

```python
#Evaluating the model
```

```python
y_predict = regressor.predict(X_test)
y_predict
```

```python
### visualise the Results
```

```python
plt.scatter(y_test, y_predict, color = 'r')
plt.title('Multiple Linear Regression Prediction')
plt.xlabel('True Value (Ground Truth)')
plt.ylabel('Mode of Prediction')
```

```python
k =X_test.shape[1]
n = len(X_test)
```

```python
n
```

```python
k
```

```python
# Doing Some Mathematics
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)),
                    '.3f'))

MSE =mean_squared_error(y_test, y_predict)
MAE =mean_absolute_error(y_test, y_predict)
r2=r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2) * (n-1)/(n-k-1)
MAPE = np.mean(np.abs((y_test-y_predict)/y_test))*100
```

```python
print('RMSE = ', RMSE)
print('MSE = ', MSE)
print('Adjested R2 = ',adj_r2)
print('MAPE= ', MAPE)
```
